# K-Means Clustering

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('customer_segmentation.csv')
# df.head()
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
X = df.copy()
X.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X)
X=scaler.transform(X)
```

```python
from sklearn.cluster import KMeans

k = 4
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
pd.Series(labels).value_counts()
```

```python
clusters = pd.DataFrame(X, columns=df.columns)
clusters['kmeans'] = labels
```

```python
import plotly.express as px

fig = px.scatter_3d(clusters, x='n_clicks', y='amount_spent', z='amount_discount', color='kmeans', width=600, height=600)
fig.update_traces(marker=dict(size=2), selector=dict(mode='markers'))
fig.show()
```

### Elbow Method

```python
wcss = []
for k in range(1,10):
    m = KMeans(n_clusters=k).fit(X)
    wcss.append(m.inertia_)

plt.plot(np.arange(1,10), wcss, '-o')
```

# Heirarchical Clustering

**Using Scipy - dendogram**,
**Smaller data set - demo**

```python
df_stocks = pd.read_csv('Stocks.csv', index_col=0)
```

```python
df_stocks.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_stocks)
X_stock = scaler.transform(df_stocks)
```

```python
scaled_df = pd.DataFrame(X_stock, columns=df_stocks.columns, index=df_stocks.index)
```

```python
import scipy.cluster.hierarchy as sch

Z = sch.linkage(scaled_df, metric='euclidean', method='ward')
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
sch.dendrogram(Z, labels=scaled_df.index, ax=ax, color_threshold=2)
plt.xticks(rotation=90)
ax.set_ylabel('distance')
```

**Our case**, 
**Using Sklearn -- industry way, no vis support**

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'ward')
y_pred = hc.fit_predict(X)
```

```python
clusters['hc'] = y_pred
```

```python
import plotly.express as px

fig = px.scatter_3d(clusters, x='n_clicks', y='amount_spent', z='amount_discount', color='hc', width=600, height=600)
fig.update_traces(marker=dict(size=2), selector=dict(mode='markers'))
fig.show()
```
